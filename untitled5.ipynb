```python
import numpy as np
import matplotlib.pyplot as plt


try:
    import myode
except Exception:
    myode = None



def diffeq(y):
    x, vx, yy, vy = y
    r = np.sqrt(x**2 + yy**2)
    ax = -x / (r**3)
    ay = -yy / (r**3)
    return np.array([vx, ax, vy, ay])


def euler_step(y, dt):
    return y + dt * diffeq(y)

def rk2_step(y, dt):
    k1 = diffeq(y)
    y_mid = y + 0.5 * dt * k1
    k2 = diffeq(y_mid)
    return y + dt * k2

def energy(y):
    x, vx, yy, vy = y
    r = np.sqrt(x**2 + yy**2)
    K = 0.5 * (vx**2 + vy**2)
    U = -1.0 / r
    return K + U

def angular_momentum(y):
    x, vx, yy, vy = y
    return x * vy - yy * vx
```

```python
#Part 1 - Circular Orbit

dt = 0.01
T_orbit = 2 * np.pi * 1.0**(3/2)
t_range = np.arange(0, 2*T_orbit, dt)


r0 = 1.0
vc = 1.0 / np.sqrt(r0)
y0 = np.array([r0, 0.0, 0.0, vc])


xs, ys, ts = [], [], []
y = y0.copy()

for t in t_range:
    xs.append(y[0])
    ys.append(y[2])
    ts.append(t)
    y = rk2_step(y, dt)

xs = np.array(xs)
ys = np.array(ys)
ts = np.array(ts)


fig, axes = plt.subplots(1, 2, figsize=(10, 4))


axes[0].plot(xs, ys)
axes[0].plot([0], [0], 'ko', markersize=4)
axes[0].set_aspect('equal')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Orbit in the x-y plane (RK2)')


axes[1].plot(ts, xs, label='x(t)')
axes[1].plot(ts, ys, label='y(t)')
axes[1].set_xlabel('time')
axes[1].set_ylabel('position')
axes[1].set_title('x(t) and y(t)')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
#The orbit isn't perfect, but its close. Since RK2 is an approximation, there's going to be a numerical error
#If I'd increase dt a lot, the orbit will drift off.
#With large time steps, it'll become less accurate, so numerical errors get bigger. This will cause the orbit to change size
```

```python
#Part 2
```

```python
dt = 0.2
T_orbit = 2* np.pi
t_range = np.arange(0, 5*T_orbit, dt)
y0 = np.array([1.0, 0.0, 0.0, 1.0])

xs_euler, ys_euler = [], []
y = y0.copy()
for t in t_range:
    xs_euler.append(y[0])
    ys_euler.append(y[2])
    y = euler_step(y, dt)


xs_rk2, ys_rk2 = [], []
y = y0.copy()
for t in t_range:
    xs_rk2.append(y[0])
    ys_rk2.append(y[2])
    y = rk2_step(y, dt)

xs_euler = np.array(xs_euler); ys_euler = np.array(ys_euler)
xs_rk2 = np.array(xs_rk2);     ys_rk2 = np.array(ys_rk2)

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(xs_euler, ys_euler, label='Euler')
ax.plot(xs_rk2, ys_rk2, label='RK2')
ax.plot([0], [0], 'ko', markersize=4)
ax.set_aspect('equal')
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_title('Euler vs RK2, dt=0.2')
ax.legend()
plt.show()
```

```python
#The orbit stops looking like a stable circle and blew up into a spiral using Euler's Method. RK2 still looks like an orbit
#This is becaause Euler has a larger local error. Each step has more error, and those errors build upon each other over time more than RK2
```

```python
#Part 3
```

```python
dt = 0.2
t_range = np.arange(0, 5*2*np.pi, dt)
y0 = np.array([1.0, 0.0, 0.0, 1.0])
E0 = energy(y0)
L0 = angular_momentum(y0)

E_euler, L_euler = [], []
y = y0.copy()
for t in t_range:
    E_euler.append(energy(y))
    L_euler.append(angular_momentum(y))
    y = euler_step(y, dt)


E_rk2, L_rk2 = [], []
y = y0.copy()
for t in t_range:
    E_rk2.append(energy(y))
    L_rk2.append(angular_momentum(y))
    y = rk2_step(y, dt)

E_euler = np.array(E_euler); E_rk2 = np.array(E_rk2)
L_euler = np.array(L_euler); L_rk2 = np.array(L_rk2)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t_range, E_euler, label='Euler E(t)')
ax.plot(t_range, E_rk2, label='RK2 E(t)')
ax.axhline(E0, color='k', linestyle='--', label='Initial E')
ax.set_xlabel('time'); ax.set_ylabel('Total energy E')
ax.legend(); plt.tight_layout(); plt.show()

frac_E_err_euler = (E_euler[-1] - E0) / E0
frac_E_err_rk2   = (E_rk2[-1] - E0) / E0
print("Fractional energy error (final vs initial):")
print(f"  Euler: {frac_E_err_euler:.4e}")
print(f"  RK2:   {frac_E_err_rk2:.4e}")

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t_range, L_euler, label='Euler L(t)')
ax.plot(t_range, L_rk2, label='RK2 L(t)')
ax.axhline(L0, color='k', linestyle='--', label='Initial L')
ax.set_xlabel('time'); ax.set_ylabel('Angular momentum L')
ax.legend(); plt.tight_layout(); plt.show()

frac_L_err_euler = (L_euler[-1] - L0) / L0
frac_L_err_rk2   = (L_rk2[-1] - L0) / L0
print("Fractional angular momentum error (final vs initial):")
print(f"  Euler: {frac_L_err_euler:.4e}")
print(f"  RK2:   {frac_L_err_rk2:.4e}")
```

```python
#Energy will drift more for Euler's. The energy should increase, which means orbit gains energy numerically and the planet spirals outward
#RK2 has some drift still, but it's smaller
#In the real two body problem, total energy should be conserved. If the numerical method changes energy that shows the method/time step is not accurate
```

```python
#Part 4
```

```python
radii = [0.5, 1.0, 2.0, 4.0]
measured_periods = []

for r0 in radii:
    vc = 1.0 / np.sqrt(r0)
    y0 = np.array([r0, 0.0, 0.0, vc])
    T_theory = 2 * np.pi * r0**(3/2)
    dt = T_theory / 500
    t_range = np.arange(0, 2*T_theory, dt)

    y = y0.copy()
    prev_x = y[0]
    crossed_below = False
    period = None

    for t in t_range[1:]:
        y = rk2_step(y, dt)
        x = y[0]

        if x < r0:
          crossed_below = True


        if crossed_below and (prev_x < r0) and (x >= r0):
            period = t
            break

        prev_x = x

    measured_periods.append(period)
    print(f"r={r0}: T_measured={period:.4f}, T_theory={T_theory:.4f}")

r_arr = np.array(radii)
T_arr = np.array(measured_periods)
r_fine = np.linspace(0.4, 5, 100)

fig, ax = plt.subplots(figsize=(5, 4))
ax.loglog(r_arr, T_arr, 'bo', label='Measured')
ax.loglog(r_fine, 2*np.pi*r_fine**(3/2), 'k--', label='Theory $T=2\\pi r^{3/2}$')
ax.set_xlabel('Radius r'); ax.set_ylabel('Period T')
ax.legend(); plt.tight_layout(); plt.show()

slope = np.polyfit(np.log(r_arr), np.log(T_arr), 1)[0]
print(f"Measured exponent: {slope:.3f}  (expected: 1.500)")




```

```python
#The measured exponent is close to 3/2, which is pretty close to Kepler's third law
#If there's a discrepency, probably a numerical reasondue to the finite timestep.
```

```python
#Part 5
```

```python
dt = 0.01
t_end = 20 * np.pi
t_range = np.arange(0, t_end, dt)

y0 = np.array([1.0, 0.0, 0.0, 0.8])
y = y0.copy()

xs, ys = [], []
Es, Ls = [], []
for t in t_range:
    xs.append(y[0]); ys.append(y[2])
    Es.append(energy(y))
    Ls.append(angular_momentum(y))
    y = rk2_step(y, dt)

xs = np.array(xs); ys = np.array(ys)
Es = np.array(Es); Ls = np.array(Ls)

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(xs, ys)
ax.plot([0], [0], 'ko', markersize=4)
ax.set_aspect('equal')
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_title('Elliptical orbit (RK2)')
plt.show()

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t_range, Es)
ax.axhline(energy(y0), color='k', linestyle='--', label='Initial E')
ax.set_xlabel('time'); ax.set_ylabel('E')
ax.set_title('Energy vs time (elliptical orbit)')
ax.legend()
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t_range, Ls)
ax.axhline(angular_momentum(y0), color='k', linestyle='--', label='Initial L')
ax.set_xlabel('time'); ax.set_ylabel('L')
ax.set_title('Angular momentum vs time (elliptical orbit)')
ax.legend()
plt.tight_layout()
plt.show()

E0 = energy(y0)
L0 = angular_momentum(y0)
print("Part 5 drift check:")
print(f"  Energy change: {(Es[-1]-E0)/E0:.4e}")
print(f"  Angular momentum change: {(Ls[-1]-L0)/L0:.4e}")
```

```python
#The orbit is elliptical. Energy and angular momentum are relatively conserved. There is a small drift because RK2 is still an approximation.
```
