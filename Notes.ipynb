```python
#loop of an object faling without air resistance
y = 100 #initial height of object in meters
g = -9.81
v = 0
m = 4
p = m*v
deltat = 0.001
while y >= 0:
  print("Object's height is: ", y)
  print("Object's velocity downwards is: ", abs(v))
  p += m*g*deltat
  v = p/m
  y += v*deltat
```
