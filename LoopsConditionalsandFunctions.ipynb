```python
for n in range(1, 101):
  prime = True

  if n <= 1:
    prime = False
  else:
    for d in range(2, n):
      if n % d == 0:
        prime = False

  if prime == True:
    print(n)
```

```python
import numpy as np

def forward_derivative(f, x, h):
  return (f(x+h) - f(x)) / h

def poly(x):
  return x**4 - 2*x**3 + x

def trig(x):
  return np.sin(x)

def expo(x):
  return np.exp(x)

x=1.0
h_values = [0.1, 0.01, 0.001]

for h in h_values:
  print("h =", h)

  print("Polynomial derivative:", forward_derivative(poly, x, h))
  print("sin(x) derivative:", forward_derivative(trig, x, h))
  print("exp(x) derivative:", forward_derivative(expo, x, h))

  print()
```
