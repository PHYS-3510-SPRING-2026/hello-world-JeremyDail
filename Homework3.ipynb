```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 20.01, 0.01)

def f(t):
  return np.sin(4*t) * np.cos(4*t)

def g(t):
  y = np.sin(t) / t
  y[t==0] = 1
  return y

def h(t):
  return np.tan(2*t)

tp = np.arange(-10, 10.01, 0.01)

fp = f(tp)
gp = g(tp)
hp = h(tp)

plt.figure(figsize = (10,6))

plt.plot(tp, fp, label = 'f(t) = sin(4t)cos(4t)')
plt.plot(tp, gp, label='g(t) = sin(t)/t')
plt.plot(tp, hp, label='h(t) = tan(2t)')

k=np.arange(-20, 20)
asymptotes = np.pi/4+k*(np.pi/2)

for a in asymptotes:
  if -10 < a < 10:
    plt.axvline(a,linestyle ='--')

plt.xlim(-10,10)
plt.xlabel('t (seconds)')
plt.ylabel('Function value')
plt.title('Plots of f(t), g(t), and h(t)')
plt.legend()
plt.grid(True)

plt.show()
```
